```python
A = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
B = A[::-1]
C = A[:, ::-1]

D = np.stack((A, B, C))
print(D.shape)
D
```

```python
A = np.ones((2, 1, 2))
B = np.zeros((2, 3, 2))

C = np.concatenate((A, B), 1)
print(C.shape)
C
```
